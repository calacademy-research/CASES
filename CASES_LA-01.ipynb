## CASES, Los Angeles
The following is a worked example of the model using data from Los Angeles. The data include unemployment figures from March 2020, and an estimate of mortality rate based on the first 100 days since the first fatality in Los Angeles. The unemployment numbers between March and April were used to estimate unemployment rates for each sector.

```julia
using DifferentialEquations, Plots, DelimitedFiles, CSV
```

```julia
# reading a delimited file
reqs = readdlm("US_exchanges_2018.csv")
```

```julia
# transpose the USBLS array
I = reqs'
# sum the rows
sum_I = sum(I, dims=2)
# normalize elements in row according to row sums
# therefore total support for each sector sums to 1
no_sectors = size(I,1)
B = Array{Float64}(undef,no_sectors,no_sectors)
for i = 1:no_sectors
    for j = 1:no_sectors
        B[i,j] = I[i,j]/sum_I[i]
    end
end
print(no_sectors)
```

The threshold for recovery above is t=167. Beyond that, the differential system is unsolvable. Figured out why, and it's a "duh". The system is a "compartmental" model, in this case the compartments being employed, unemployed, and removed (dead). The system is now well-behaved. Recovery is also a threshold affair, and the good news here is that it occurs much more rapidly than collapse does! Therefore, if recovery was simply a reversal of decline, we could expect to see a return to full employment in a fraction of the time that it took for the system to crash. Realistically, however, the would be extremely difficult given that you are asking the economy to grow after it has collapsed. This is where external stimulus could be very important.

```julia
s2 = Array(sol4);
e = Float64[]
for i = 1:size(s2,2)
    e1 = 0
    for j = 1:no_sectors
        e1 = e1 + s2[j,i]
    end
    push!(e,e1)
end

d_e = Float64[]
for i = 1:size(e,1)-no_sectors
    push!(d_e, abs(e[i+1]-e[i])/(e[i+2]-e[i+3]))
end
for i = 1:size(d_e,1)-1
    if d_e[i+1]/d_e[i] >= 2.0
        println(i+1," ",e[i+1])
        break
    end
end
```
